# Baseline

*Train config:* TrainPhysTks-EvalPhysTks-NoReweight <br>
*Net config:* TMVA-Run2-NoTkLikCDVelodEdx <br>
*GhostAccFrac:* 1.0 <br>
*Training data:* Mixture <br>
*Data files:* MC2015Sim09Dev03 <br>
